```python
df_suffixed = window_df.add_suffix('_df')
pred_df_suffixed = pred_df.add_suffix('_pred_df')
df = df_suffixed.join(pred_df_suffixed, how='outer')
df['lag_time_pred_df'].dropna()
```

```python
def backtest(df):
    capital = 1000
    trading_fees_pct = 0  # Trading fees percentage
    trade_results = []  # To store results of each trade

    # Iterate over rows in the DataFrame
    for i in range(len(df)):
        # Check if met_threshold_pred_df is 0, lag_time_pred_df is 0 or NaN, skip the trade
        if df['met_threshold_pred_df'].iloc[i] == 0 or pd.isna(df['lag_time_pred_df'].iloc[i]) or df['lag_time_pred_df'].iloc[i] == 0:
            continue

        # If lag_time_pred_df would cause us to go beyond the DataFrame, skip the trade
        lag_time = int(df['lag_time_pred_df'].iloc[i])  # Ensure lag_time is an integer
        if i + lag_time >= len(df):
            continue

        # Calculate the investment amount (50% of the current capital)
        investment = capital * 0.5

        # Get the start and end price
        start_price = df['ADA_close_df'].iloc[i]
        end_price = df['ADA_close_df'].iloc[i + lag_time]

        # Calculate percentage change in price
        pct_change = (end_price - start_price) / start_price
        
        # Calculate profit/loss
        profit_loss = investment * pct_change

        # Apply trading fees
        profit_loss -= investment * trading_fees_pct

        # Update capital
        capital += profit_loss

        # Store the trade result
        trade_results.append({
            'start_price': start_price,
            'end_price': end_price,
            'pct_change': pct_change,
            'profit_loss': profit_loss,
            'capital_after_trade': capital
        })

    # Convert trade_results to a DataFrame
    trade_results_df = pd.DataFrame(trade_results)
    return trade_results_df
```

```python
backtest(df)
```
