```python
def portfolio_plotter(df, ticker):
    df = pd.read_csv(f'../data/portfolio-performance/{ticker}_portfolio.csv').drop('Unnamed: 0', axis = 1)
    # Ensure 'open_time' and 'close_time' are datetime
    df['open_time'] = pd.to_datetime(df['open_time'])
    df['close_time'] = pd.to_datetime(df['close_time'])

    # Create an empty DataFrame spanning the entire time range at one minute intervals
    new_index = pd.date_range(start=df['open_time'].min(), end=df['close_time'].max(), freq='T')
    new_df = pd.DataFrame(index=new_index)

    # For each row in the original DataFrame, forward fill 'capital' in the new DataFrame
    for idx, row in df.iterrows():
        new_df.loc[row['open_time']:row['close_time'], 'capital'] = row['capital']

    # Forward fill the 'capital' values in the new DataFrame
    new_df['capital'].ffill(inplace=True)

    # Plot the data
    fig, ax = plt.subplots(figsize=(15, 10)) 
    new_df['capital'].plot(ax=ax, style='-o', color='darkblue', label='capital')

    # Add a horizontal line at y=1000
    ax.axhline(1000, color='red', linestyle='--', label='Baseline = 1000')

    # Set the title and labels
    ax.set_title(f'{ticker} Portfolio Over Time', fontsize=20)
    ax.set_xlabel('Time', fontsize=16)
    ax.set_ylabel('Capital', fontsize=16)

    # Improve the aesthetics
    ax.grid(True) # Add gridlines
    plt.xticks(rotation=45) # Rotate x-axis labels for better visibility
    ax.legend() # Add a legend
    plt.savefig(f'../images/{ticker}_portfolio')

    return plt.show()
```

```python
cryptos = ['ADA', 'BNB', 'DOGE', 'ETH', 'LINK', 'LTC', 'UNI', 'XRP']
for crypto in cryptos:
    portfolio_plotter(df, ticker = crypto)
```

```python
df = pd.read_csv('../data/merged_data/all_currencies.csv')
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# Increase default font size
mpl.rcParams['font.size'] = 14

# Calculate the total volume for each cryptocurrency
total_volumes = df.filter(regex='_volume$').sum()

# Generate labels from the column names, removing the '_volume' part
labels = total_volumes.index.str.replace('_volume', '')

# Sort volumes and labels by volume size
total_volumes, labels = zip(*sorted(zip(total_volumes, labels)))

# Create a function to label percentages over a certain threshold
def autopct_format(values):
    def my_format(pct):
        return ('%1.1f%%' % pct) if pct > 4 else ''
    return my_format

# Generate the pie chart
plt.figure(figsize=(12, 10))
wedges, texts, autotexts = plt.pie(total_volumes, labels=labels, autopct=autopct_format(total_volumes), 
                                    startangle=140, shadow=True, pctdistance=0.85)

# Draw a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Set the autotext color to white
plt.setp(autotexts, color='white', weight="bold")

plt.title("Total Volume for Each Cryptocurrency", weight="bold", size=20)
plt.savefig('../images/Volume_pie_chart')
plt.show()
```

```python
df['time'] = pd.to_datetime(df['time'])

# Plot ADA_open price against time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='time', y='ADA_open', marker='o', markersize=4)

# Find the index of the maximum ADA_open value
max_index = df['ADA_open'].idxmax()

# Add a vertical line at the maximum value
plt.axvline(x=df['time'][max_index], color='red', linestyle='--', label='Max ADA_open')

# Set labels and title
plt.xlabel('Time')
plt.ylabel('ADA Price')
plt.title('ADA Price vs. Time with Max Value Highlighted')
plt.legend()

# Show the plot
plt.tight_layout()
plt.savefig('../images/window_graph')
plt.show()
```
