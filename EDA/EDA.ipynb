```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
df = pd.read_csv("../data/ready/ADA.csv", index_col='time').drop('Unnamed: 0', axis = 1)
df.dropna().head(50)
```

```python
df.index = pd.to_datetime(df.index)

fig, ax = plt.subplots(figsize=(15, 10)) 

df['BTC_open'].plot(ax=ax, style='-o', label='BTC 5% Jump')

df.loc[df['lag_time'].notna(), 'BTC_open'].plot(ax=ax, style='ro', label='lag time not null')

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

ax.set_title('Bitcoin Opening Price Over Time', fontsize=20)
ax.set_xlabel('Time', fontsize=16)
ax.set_ylabel('BTC Open Price', fontsize=16)

plt.xticks(rotation=45)

plt.legend()

plt.show()
```

```python
df = pd.read_csv('../data/portfolio-performance/ADA_portfolio.csv').drop('Unnamed: 0', axis = 1)
```

```python
def portfolio_plotter(df):
    # Ensure 'open_time' and 'close_time' are datetime
    df['open_time'] = pd.to_datetime(df['open_time'])
    df['close_time'] = pd.to_datetime(df['close_time'])

    # Create an empty DataFrame spanning the entire time range at one minute intervals
    new_index = pd.date_range(start=df['open_time'].min(), end=df['close_time'].max(), freq='T')
    new_df = pd.DataFrame(index=new_index)

    # For each row in the original DataFrame, forward fill 'capital' in the new DataFrame
    for idx, row in df.iterrows():
        new_df.loc[row['open_time']:row['close_time'], 'capital'] = row['capital']

    # Forward fill the 'capital' values in the new DataFrame
    new_df['capital'].ffill(inplace=True)

    # Plot the data
    fig, ax = plt.subplots(figsize=(15, 10)) 
    new_df['capital'].plot(ax=ax, style='-o', color='darkblue', label='capital')

    # Add a horizontal line at y=1000
    ax.axhline(1000, color='red', linestyle='--', label='Baseline = 1000')

    # Set the title and labels
    ax.set_title('Capital Over Time', fontsize=20)
    ax.set_xlabel('Time', fontsize=16)
    ax.set_ylabel('Capital', fontsize=16)

    # Improve the aesthetics
    ax.grid(True) # Add gridlines
    plt.xticks(rotation=45) # Rotate x-axis labels for better visibility
    ax.legend() # Add a legend

    return plt.show()
```

```python
portfolio_plotter(df)
```
