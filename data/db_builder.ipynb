```python
from pathlib import Path
import sqlite3
import pandas as pd
```

```python
Path('crypto_database.db').touch()
conn = sqlite3.connect('crypto_database.db')
c = conn.cursor()
```

```python
BTC = pd.read_csv('raw/BTC_price_data.csv')
BTC.to_sql('BTC', conn, if_exists='append', index = False)
```

```python
csv_files = {
    "ETH": "raw/ETH_price_data.csv",
    "XRP": "raw/XRP_price_data.csv",
    "LTC": "raw/LTC_price_data.csv",
    "ADA": "raw/ADA_price_data.csv",
    "DOT": "raw/DOT_price_data.csv",
    "LINK": "raw/LINK_price_data.csv",
    "BNB": "raw/BNB_price_data.csv",
    "UNI": "raw/UNI_price_data.csv",
    "DOGE": "raw/DOGE_price_data.csv"
}
```

```python
for symbol, file_path in csv_files.items():
    c = conn.cursor()
    c.execute(f'''CREATE TABLE {symbol} (time TEXT, high REAL, low REAL, open REAL, volumefrom REAL, volumeto REAL, close REAL, conversionType TEXT, conversionSymbol TEXT)''')
    df = pd.read_csv(file_path)
    df.to_sql(symbol, conn, if_exists='append', index = False)
```
